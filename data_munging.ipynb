#### Clean baby name df function
* Assign column names
* Drop the number of babies contained in column `xyz` 
* Change genders to binary
    * `F` $\rightarrow$ `1`
    * `M` $\rightarrow$ `0`

```python
def clean_baby_df(df):
    df.columns = ['name', 'gender', 'xyz']
    df.drop('xyz', axis=1, inplace=True)
    df.gender.replace({'F': 1, 'M': 0})
    return df
```
